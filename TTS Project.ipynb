```python
# 3. ADIM: SES SENTEZİ KODU (YENİ RUNTIME'DA ÇALIŞTIRIN)
from TTS.api import TTS
import IPython.display as ipd
import scipy.io.wavfile as wavfile
from IPython.display import display, Audio
import numpy as np

# Kullanılabilir modelleri listele
print("Mevcut TTS modelleri:")
print(TTS().list_models())

# Türkçe modelini yükle
try:
    tts = TTS(model_name="tts_models/tr/common-voice/glow-tts", progress_bar=True, gpu=False)
    print("Türkçe model başarıyla yüklendi!")
except Exception as e:
    print(f"Türkçe model yüklenemedi: {e}")
    print("İngilizce model kullanılıyor...")
    tts = TTS(model_name="tts_models/en/ljspeech/glow-tts", progress_bar=True, gpu=False)

# Metni seslendirme
text = "Merhaba, ben Coqui TTS ile Türkçe metin okuyorum."
print(f"Seslendiriliyor: '{text}'")
wav = tts.tts(text)

# Çıktıyı numpy array'e dönüştürme
if isinstance(wav, list):
    wav = np.array(wav)

# Ses dosyasını kaydetme
sample_rate = 22050
wavfile.write("output.wav", sample_rate, wav)

# Ses dosyasını çalma
print("Ses çalınıyor...")
display(Audio("output.wav", autoplay=True))

# Ek bilgi
print("\nİşlem tamamlandı! 'output.wav' dosyası oluşturuldu.")
```

```python
import time

# Metni seslendirme ve süreyi ölçme
start_time = time.time()
text = "Merhaba, ben Coqui TTS ile Türkçe metin okuyorum."
wav = tts.tts(text)
end_time = time.time()

# Geçen süreyi hesapla
duration = end_time - start_time  # saniye cinsinden süre
rtf = duration / len(text)  # RTF hesaplama
latency = duration  # Latency de zaten geçiş süresiyle aynı olacak

print(f"RTF (Real-Time Factor): {rtf:.4f}")
print(f"Latency (Gecikme): {latency:.4f} saniye")
```
